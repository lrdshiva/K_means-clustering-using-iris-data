K means clustering is an unsupervised learning algorithm which groups unclassified data into clusters. 
Refer to the link to know more https://www.datascience.com/blog/k-means-clustering

Other useful links:

```
#data dependencies
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import numpy as np

iris_data = datasets.load_iris()
X = iris_data.data
y = iris_data.target
```

```
X.shape
```

Col 1 : sepal length , Col 2 : sepal width
Col 3 : petal length , Col 4 : petal width

```
for i in range(0,150):
  print ("Features ",X[i,:]," label =",y[i] )
```

#visualization
For visualizing data in 3D we need to used PCA as there are 4 Features

http://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html

```

PCA1 = PCA(n_components = 3) #reduce from 4 dimensions to 3
X_PCA = PCA1.fit_transform(X)

#Fix the dimension of the figure
fig = plt.figure(1, figsize=(8, 6))
ax = Axes3D(fig)
ax.scatter(X_PCA[:, 0], X_PCA[:, 1], X_PCA[:, 2], c=y, cmap=plt.cm.Set1, edgecolor='k', s=40)


ax.set_title(" PCA directions")
ax.set_xlabel("1st eigenvector")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("2nd eigenvector")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3rd eigenvector")
ax.w_zaxis.set_ticklabels([])
```

```
X_PCA.shape
```

#model creation 

```
Model= KMeans(n_clusters = 3)  # I'll try with different n_init and max_iter

Model.fit(X)

```

```
label = Model.labels_
label
```

```
centers = Model.cluster_centers_ 
centers
```

###label matching
Here we see that the flowers have been labeled in different way.
so we try to match the labels. In the K means model

----------------True label------model label

Setosa----------------------0                    1

Versicolour---------------1                    0

Virginica-------------------2                    2



```

y_changed = np.choose(label, [1, 0 , 2])
print(y_changed)
```

```
for i in range(0,150):
  print("True label = ",y[i]," Kmean predicted value = ", y_changed[i])
```

#plot using predicted value

```
PCA2 = PCA(n_components = 3) #reduce from 4 dimensions to 3
X_PCA_predicted = PCA2.fit_transform(X)

#Fix the dimension of the figure
fig = plt.figure(1, figsize=(8, 6))
ax = Axes3D(fig)
ax.scatter(X_PCA_predicted[:, 0],
           X_PCA_predicted[:, 1],
           X_PCA_predicted[:, 2],
           c=y_changed,
           cmap=plt.cm.Set1
           , edgecolor='k', s=40)
  
  
PCA3 = PCA(n_components = 3) #reduce from 4 dimensions to 3
X_PCA_centers = PCA3.fit_transform(centers)



ax.text3D(X_PCA_centers[0,0],
          X_PCA_centers[0,1],
          X_PCA_centers[0,2],
          "Setosa",
          horizontalalignment='center',
          bbox=dict(alpha=.2, edgecolor='w', facecolor='w'))

ax.text3D(X_PCA_centers[1,0],
          X_PCA_centers[1,1],
          X_PCA_centers[1,2],
          "Versicolour",
          horizontalalignment='center',
          bbox=dict(alpha=.2, edgecolor='w', facecolor='w'))

ax.text3D(X_PCA_centers[2,0],
          X_PCA_centers[2,1],
          X_PCA_centers[2,2],
          "Virginica",
          horizontalalignment='center',
          bbox=dict(alpha=.2, edgecolor='w', facecolor='w'))


ax.set_title(" PCA directions for predicted value")
ax.set_xlabel("1st eigenvector")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("2nd eigenvector")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3rd eigenvector")
ax.w_zaxis.set_ticklabels([])
```

#Accuracy

```
from sklearn.metrics import accuracy_score

print('Accuracy: %.2f' % accuracy_score(y, y_changed))
```
